```
import matplotlib.pyplot as plt
```

```
plt.figure(figsize=(10,8))
```

```
counts = df['num_failed_logins'].value_counts()

plt.figure(figsize=(10,8))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Distribution of Failed Logins')
plt.show()
```

```
counts = df['num_failed_logins'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(counts.index, counts)
plt.xlabel('Number of Failed Logins')
plt.ylabel('Count')
plt.title('Distribution of Failed Logins')
plt.show()
```

```
counts = df['num_failed_logins'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.plot(counts.index, counts, marker='o')
plt.xlabel('Number of Failed Logins')
plt.ylabel('Count')
plt.title('Distribution of Failed Logins (Line Plot)')
plt.grid(True)
plt.show()
```

```
import matplotlib.pyplot as plt

counts = df['num_failed_logins'].value_counts()
counts_sorted = counts.sort_index()


plt.figure(figsize=(20, 6))


plt.subplot(1, 3, 1)
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Failed Logins Distribution (Pie Chart)')


plt.subplot(1, 3, 2)
plt.bar(counts.index, counts)
plt.xlabel('Number of Failed Logins')
plt.ylabel('Count')
plt.title('Failed Logins Distribution (Bar Chart)')
plt.grid(axis='y', linestyle='--', alpha=0.6)


plt.subplot(1, 3, 3)
plt.plot(counts_sorted.index, counts_sorted, marker='o', linewidth=2)
plt.xlabel('Number of Failed Logins')
plt.ylabel('Count')
plt.title('Failed Logins Distribution (Line Chart)')
plt.grid(True, linestyle='--', alpha=0.6)


plt.tight_layout()
plt.savefig('failed_login_distribution_dashboard.png', dpi=300, bbox_inches='tight')

plt.show()
```
